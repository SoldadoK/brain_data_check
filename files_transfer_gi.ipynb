Steps:
1. Identify and select files based on date (modified) and formats (edat3 and txt. and xml. file). Be alert: the weight of the file could be different. 
2. Display selected files. 
3. Transfers all selected edat3 files to the map: \raw
4. Transfer all selected txt and xml files to the map: \source
5. Display final overview/table: number of files transfered to the map raw, list all of them (write the full name of the file);
number of files transfered to the map source, list all of them (write the full name of the file). 

```python
import os
import shutil
from datetime import datetime

# Source directory from where the files will be transferred
source_directory = '/export_lab-visit'

# File formats we are looking for
file_formats = ['edat3', 'txt', 'xml']

# Define a date for filtering files modified on or after this date. 
#Example: date_string = "22.08.2023"
date_string = "17.08.2023"
day, month, year = map(int, date_string.split('.'))
date_filter = datetime(year, month, day)

# List to hold selected files
selected_files = {
    'edat3': [],
    'txt': [],
    'xml': []
}

# Iterate through the files in the source directory and filter based on the formats and date
for file_name in os.listdir(source_directory):
    if file_name.split('.')[-1] in file_formats:
        file_path = os.path.join(source_directory, file_name)
        file_modified_date = datetime.fromtimestamp(os.path.getmtime(file_path))
        
        if file_modified_date >= date_filter:
            selected_files[file_name.split('.')[-1]].append(file_path)

# Print the selected files
print("Selected files:")
for format, files in selected_files.items():
    print(f"{format}: {len(files)} files")
    for file in files:
        print(f"  - {os.path.basename(file)}")

# Destination directories
destination_raw = '/raw'
destination_source = '/source'

# Transfer edat3 files to raw directory
for file_path in selected_files['edat3']:
    shutil.move(file_path, os.path.join(destination_raw, os.path.basename(file_path)))

# Transfer txt and xml files to source directory
for file_format in ['txt', 'xml']:
    for file_path in selected_files[file_format]:
        shutil.move(file_path, os.path.join(destination_source, os.path.basename(file_path)))

# Display final overview
print("Final Overview:")
print(f"Number of files transferred to the map raw: {len(selected_files['edat3'])}")
print("List of them:")
for file in selected_files['edat3']:
    print(f"  - {os.path.basename(file)}")
print(f"Number of files transferred to the map source: {len(selected_files['txt']) + len(selected_files['xml'])}")
print("List of them:")
for file_format in ['txt', 'xml']:
    for file in selected_files[file_format]:
        print(f"  - {os.path.basename(file)}")
```
